# **Métodos y cálculos**

• Índice de temperatura y humedad (producción intensiva)

$$ITH = 0.72(T_{bs} + T_{bh}) + 40.6$$

Donde:


$T_{bs}$ Temperatura bulbo seco ($^oC$)


$T_{bh}$ Temperatura de bulbo húmedo ($^oC$)

Para la correlación de los datos de la especie bovina en la producción de leche se recomienda
utilizar los límites propuestos por National Weather Service:

Sí $ITH ≤ 70$ ambiente no estresante

Sí $ITH > 70 \space And \space ITH ≤ 78$ situación de alerta por estrés térmico

Sí $ITH > 78 \space And \space ITH ≤ 83$ peligro por estrés térmico

Sí $ITH > 83$ situación crítica





````python
# ENTRADA: Pedir tbs (Temperatura de bulbo secoo) y tbh (Temperatura de bulbo húmedo)
tbs = float(input("Ingrese la temperatura de bulbo seco, Grados Centigrados: "))
tbh = float(input("Ingrese la temperatura de bulbo húmedo, Grados Centigrados: "))

# PROCESO: Calculo de ith (índice de Temperatura y Humedad)
ith = 0.72 * (tbs + tbh) + 40.6

# SALIDA: Imprimir resultado de la evaluación del ith y el confort térmico
if ith <= 70:
  print("El índice de temperatura y humedad es: ", ith)
  print("Ambiente no estresante")

elif ith > 70 and ith <= 78:
  print("El índice de temperatura y humedad es: ", ith)
  print("Situación de alerta por estrés térmico")

elif ith > 78 and ith <= 83:
  print("El índice de temperatura y humedad es: ", ith)
  print("peligro por estrés térmico")

elif ith > 83:
  print("El índice de temperatura y humedad es: ", ith)
  print("Situación crítica")
````

## **Teoría del Confort Térmico en Ganado Vacuno**

El confort térmico en el ganado vacuno se refiere a la condición en la cual el animal no necesita activar mecanismos fisiológicos o de comportamiento para mantener su temperatura corporal interna dentro de un rango normal. Cuando el ganado se encuentra fuera de su zona de termoneutralidad, experimenta estrés térmico, lo que puede tener efectos negativos significativos en su salud, productividad y bienestar.

**Factores que Afectan el Confort Térmico:**

1.  **Temperatura del aire:** Es el factor más obvio y directo. Temperaturas elevadas o muy bajas pueden inducir estrés.
2.  **Humedad relativa:** La alta humedad dificulta la disipación de calor por evaporación (sudoración y jadeo), haciendo que el ganado sienta más calor de lo que indicaría solo la temperatura del aire. Este es el motivo por el cual se utilizan índices como el ITH (Índice de Temperatura y Humedad).
3.  **Radiación solar:** La exposición directa al sol aumenta la carga de calor en los animales.
4.  **Velocidad del viento:** Un flujo de aire adecuado puede ayudar a la disipar el calor corporal, mientras que vientos muy fríos pueden agravar el estrés por frío.
5.  **Precipitación:** La lluvia o la nieve pueden alterar la temperatura corporal y el aislamiento del pelaje.

**Efectos del Estrés Térmico:**

El estrés térmico en el ganado vacuno puede llevar a:

*   **Reducción del consumo de alimento:** Los animales comen menos para generar menos calor metabólico.
*   **Disminución de la producción:** Menor producción de leche, carne y menor eficiencia reproductiva.
*   **Cambios fisiológicos:** Aumento de la frecuencia respiratoria (jadeo), aumento de la temperatura rectal, cambios hormonales.
*   **Alteraciones de comportamiento:** Búsqueda de sombra, reducción de la actividad, aumento del consumo de agua.
*   **Mayor susceptibilidad a enfermedades:** Compromiso del sistema inmunológico.

**Importancia de Monitorear el Confort Térmico:**

Monitorear y gestionar el confort térmico es crucial para la sostenibilidad y rentabilidad de la producción ganadera. Permite a los productores implementar estrategias de mitigación (sombra, ventilación, nebulización, manejo nutricional) antes de que el estrés térmico afecte gravemente a los animales, mejorando así el bienestar animal y asegurando una producción eficiente.
